```julia
sqrt_seconds(x) = try
    sqrt(x[2])
    catch y
    if isa(y, DomainError)
        sqrt(complex(x[2],0))
    elseif isa(y, BoundsError)
        sqrt(x)
    end
end
sqrt_seconds([1 4])
sqrt_seconds([1 -4])
sqrt_seconds(9)
sqrt_seconds(-9)
```

```julia
function producer()
    produce("start")
    for i=1:3:9
        produce(2i)
    end
    produce("end")
end

for x in @task producer()
    println(x)
end
```
